# Creando un grafo

```python
import networkx as nx
from networkx.readwrite import json_graph
```

```python
def formatGraphData(data):
    del data["directed"]
    del data["multigraph"]
    del data["graph"]
    return data
```

# Serializando

```python
import json
```

```python
def serializeGraph(data, filename="data.json"):
    with open(filename, 'w') as outfile:
        json.dump(data, outfile)
    print("Data saved to:", filename)
```

# Refactor

```python
def resultsTPtoJSON(results, filename="data.json"):
    G = nx.Graph()

    for result in results["results"]["bindings"]:
        t = result["t"]["value"]
        p = result["p"]["value"]
        G.add_node(str(t), color="red")
        G.add_node(str(p), color="blue")
        G.add_edge(str(t),str(p))

    data = json_graph.node_link_data(G)
    data = formatGraphData(data)
    serializeGraph(data, filename)
    print(data)
```

```python
def resultsSPOtoJSON(results, filename="data.json"):
    G = nx.Graph()

    for result in results["results"]["bindings"]:
        s = result["s"]["value"]
        o = result["o"]["value"]
        G.add_node(str(s), color="red")
        G.add_node(str(o), color="blue")
        G.add_edge(str(s),str(o))

    data = json_graph.node_link_data(G)
    data = formatGraphData(data)
    serializeGraph(data, filename)
    print(data)
```

## ?s ?p ?o

```python
start_time = time.time()

sparql = SPARQLWrapper("http://sparql.n13.degu.cl/sparql")
sparql.setQuery("""
    SELECT DISTINCT ?s ?p ?o 
    WHERE { 
        ?s ?p ?o . 
    }
    LIMIT 1000
""")
sparql.setReturnFormat(JSON)
results = sparql.query().convert()

print("Query execution: %s seconds" % round(time.time() - start_time, 2))

for result in results["results"]["bindings"]:
    print(result["s"]["value"], result["p"]["value"], result["o"]["value"])
```

# Labels

```python
start_time = time.time()

sparql = SPARQLWrapper("http://sparql.n13.degu.cl/sparql")
sparql.setQuery("""
    PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#> 
    PREFIX wd: <http://www.wikidata.org/entity/>
    PREFIX wdt: <http://www.wikidata.org/prop/direct/>
    PREFIX foaf: <http://xmlns.com/foaf/0.1/>
    
    SELECT DISTINCT ?t ?p ?l
    WHERE { 
        ?s ?p ?o .
        ?s rdf:type ?t .
        ?s rdfs:label ?l.
        FILTER (?t = foaf:Person)
        FILTER(LANG(?l) = "" || LANG(?l) = "en")
    }
""")
sparql.setReturnFormat(JSON)
results = sparql.query().convert()

print("Query execution: %s seconds" % round(time.time() - start_time, 2))

G = nx.Graph()

for result in results["results"]["bindings"]:
    s = result["s"]["value"]
    o = result["o"]["value"]
    l = result["l"]["value"]
    G.add_node(str(s), color="teal")
    G.add_node(str(o), color="purple")
    G.add_node(str(s), label=l)
    G.add_edge(str(s),str(o))

data = json_graph.node_link_data(G)
data = formatGraphData(data)
serializeGraph(data, "data.json")
```

# ?t1 ?p ?t2 

```python
start_time = time.time()

sparql = SPARQLWrapper("http://sparql.n13.degu.cl/sparql")
sparql.setQuery("""
    PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#> 
    PREFIX wd: <http://www.wikidata.org/entity/>
    PREFIX wdt: <http://www.wikidata.org/prop/direct/>
    PREFIX foaf: <http://xmlns.com/foaf/0.1/>

    SELECT DISTINCT ?t1 ?p ?t2  
    WHERE { 
        ?s ?p ?o .
        ?s rdf:type ?t1 .
        ?o rdf:type ?t2 .
        FILTER(?t1 = foaf:Person || t2 = foaf:Person)
    }
""")
sparql.setReturnFormat(JSON)
results = sparql.query().convert()

print("Query execution: %s seconds" % round(time.time() - start_time, 2))

G = nx.Graph()

for result in results["results"]["bindings"]:
    t1 = result["t1"]["value"]
    t2 = result["t2"]["value"]
    p = result["p"]["value"]
    G.add_node(str(t1), color="teal")
    G.add_node(str(t2), color="teal")
    G.add_node(str(p), color="purple")
    G.add_edge(str(t1),str(p))
    G.add_edge(str(p),str(t2))

data = json_graph.node_link_data(G)
data = formatGraphData(data)
serializeGraph(data, "data.json")
```
